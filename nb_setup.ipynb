```python
env = dbutils.widgets.get("env")
database = dbutils.widgets.get("database")
schema = dbutils.widgets.get("schema")
spark.sql(f"CREATE CATALOG IF NOT EXISTS `{env}-{database}`")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{env}-{database}.{schema}`")
```

```python
spark.sql(f"""
CREATE TABLE IF NOT EXISTS `{env}-{database}.{schema}.homes` (
  home_id STRING,
  home_name STRING,
  timezone STRING
)
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS `{env}-{database}.{schema}.sensors` (
  home_id STRING,
  device_id STRING,
  device_name STRING,
  device_type STRING,
  manufacturer STRING,
  model STRING,
  sensor_id STRING,
  sensor_name STRING,
  unit STRING,
  sensor_min DECIMAL(38,19),
  sensor_max DECIMAL(38,19),
  valid_min DECIMAL(38,19),
  valid_max DECIMAL(38,19)
)
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS `{env}-{database}.{schema}.sensor_data_raw`(
  home_id STRING,
  device_id STRING, 
  sensor_id STRING,
  reading DECIMAL(38,19),
  reading_string STRING,
  unit STRING, 
  reading_time TIMESTAMP,
  system_timestamp TIMESTAMP
) 
USING DELTA
PARTITIONED BY (home_id, reading_time)
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS `{env}-{database}.{schema}.sensor_data_cleaned`(
  home_id STRING,
  device_id STRING, 
  sensor_id STRING,
  reading DECIMAL(38,19),
  reading_string STRING,
  unit STRING, 
  reading_time TIMESTAMP,
  system_timestamp TIMESTAMP
) 
USING DELTA
PARTITIONED BY (home_id, reading_time)
""")

spark.sql(f"""
CREATE TABLE IF NOT EXISTS `{env}-{database}.{schema}.sensor_data`(
  home_id STRING,
  home_name STRING,
  timezone STRING,
  device_id STRING, 
  device_name STRING,
  device_type STRING,
  manufacturer STRING,
  model STRING,
  sensor_id STRING,
  sensor_name STRING,
  reading DECIMAL(38,19),
  reading_string STRING,
  unit STRING, 
  reading_time TIMESTAMP,
  system_timestamp TIMESTAMP
) 
USING DELTA
PARTITIONED BY (home_id, reading_time)
""")
```
